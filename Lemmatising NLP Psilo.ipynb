```python
import lemmy
import numpy as np
import re

# Create an instance of the standalone lemmatizer.
lemmatizer = lemmy.load("da")
```

```python
with open('//nrusmb3.nru.dk/data1/DrummondMcC/Language_processing/Rawdata/nonmys_wordsforlemma.txt', 'r') as f:
    words = f.readlines()
    
lwords = [None]*len(words)
for i in range(len(words)):
    words_elem = re.sub('"','',words[i].rstrip())
    lwords[i] = lemmatizer.lemmatize("", words_elem)
lwords
```

```python
words_ambig = [idx for idx in range(len(lwords)) if len(lwords[idx])>1]
lwords[words_ambig[0]]

lowrds_firstonly = [lwords[idx][0] for idx in range(len(lwords))]
len(lowrds_firstonly)
```

```python
with open('//nrusmb3.nru.dk/data1/DrummondMcC/Language_processing/Rawdata/nonmys_lemmatized_words.txt', 'w') as f:
    for item in lowrds_firstonly:
        f.write("%s\n" % item)
```
